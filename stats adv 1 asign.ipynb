```python
# 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range of the discrete uniform distribution
low = 1  # inclusive lower bound
high = 7  # exclusive upper bound (like a die roll from 1 to 6)

# Create a discrete uniform distribution object
dist = randint(low, high)

# Define x values (support of the distribution)
x = np.arange(low, high)

# Compute PMF values
pmf = dist.pmf(x)

# Plot the PMF
plt.figure(figsize=(8, 5))
plt.stem(x, pmf, basefmt=" ", use_line_collection=True)
plt.title('PMF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('x')
plt.ylabel('P(X = x)')
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# 3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x, p):
    """
    Calculate the probability mass function (PMF) of a Bernoulli distribution.
    
    Parameters:
    x (int): The outcome, which must be 0 or 1.
    p (float): The probability of success (x = 1), where 0 <= p <= 1.
    
    Returns:
    float: The probability of observing the outcome x.
    
    Raises:
    ValueError: If x is not 0 or 1 or if p is not within the interval [0, 1].
    """
    if x not in (0, 1):
        raise ValueError("The outcome x must be either 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("The probability p must be between 0 and 1 (inclusive).")
    
    return (p ** x) * ((1 - p) ** (1 - x))

# Example usage:
if __name__ == "__main__":
    p = 0.3  # Example probability of success.
    print("P(X=0):", bernoulli_pdf(0, p))  # Should print the probability of failure.
    print("P(X=1):", bernoulli_pdf(1, p))  # Should print the probability of success.
```

```python
# 6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution
import numpy as np
import matplotlib.pyplot as plt

# Define the discrete uniform distribution parameters
a = 1  # minimum value
b = 10  # maximum value

# Create the range of values
x = np.arange(a, b + 1)

# Calculate the probabilities (all equal in uniform distribution)
prob = 1 / (b - a + 1)
pmf = np.full_like(x, prob, dtype=float)

# Calculate the CDF
cdf = np.cumsum(pmf)

# Plotting
plt.figure(figsize=(8, 5))
plt.step(x, cdf, where='mid', label='CDF', color='blue')
plt.scatter(x, cdf, color='red')
plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('x')
plt.ylabel('CDF')
plt.grid(True)
plt.xticks(x)
plt.ylim(0, 1.1)
plt.legend()
plt.show()
```

```python
# 9.Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores and plot them.

    Parameters:
    - data (list or numpy array): Dataset for which Z-scores are computed.

    Returns:
    - z_scores (numpy array): Computed Z-scores.
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero. Z-scores can't be computed.")

    z_scores = (data - mean) / std

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='blue', label='Z-score')
    plt.axhline(0, color='black', linestyle='--', linewidth=1)
    plt.axhline(1, color='green', linestyle='--', linewidth=1, label='Â±1 SD')
    plt.axhline(-1, color='green', linestyle='--', linewidth=1)
    plt.axhline(2, color='orange', linestyle='--', linewidth=1, label='Â±2 SD')
    plt.axhline(-2, color='orange', linestyle='--', linewidth=1)
    plt.axhline(3, color='red', linestyle='--', linewidth=1, label='Â±3 SD')
    plt.axhline(-3, color='red', linestyle='--', linewidth=1)
    plt.title("Z-scores of Dataset")
    plt.xlabel("Index")
    plt.ylabel("Z-score")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return z_scores
```

```python
data = [10, 12, 9, 15, 20, 22, 13, 10, 8, 11]
z = calculate_and_plot_z_scores(data)
print("Z-scores:", z)
```

```python
# 10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(42)

# Step 1: Create a non-normal population (Exponential distribution)
population = np.random.exponential(scale=2.0, size=100000)

# Function to simulate sampling distribution
def simulate_clt(sample_size=30, num_samples=1000):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size)
        sample_means.append(np.mean(sample))
    return sample_means

# Step 2: Try different sample sizes
sample_sizes = [5, 30, 100]

# Plotting
plt.figure(figsize=(16, 4))

for i, n in enumerate(sample_sizes):
    means = simulate_clt(sample_size=n)
    plt.subplot(1, 3, i+1)
    sns.histplot(means, kde=True, stat="density", bins=30, color='skyblue')
    plt.title(f"Sample Size = {n}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")

plt.suptitle("Central Limit Theorem Demonstration (Exponential Population)", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# 11.Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Step 1: Set parameters
population_mean = 50
population_std = 10
sample_size = 30
num_samples = 1000

# Step 2: Generate multiple samples and calculate their means
sample_means = []

for _ in range(num_samples):
    sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
    sample_means.append(np.mean(sample))

# Step 3: Plot histogram of sample means
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(sample_means, bins=30, kde=True, stat="density", color='skyblue', label='Sample Means')

# Overlay the theoretical normal distribution
x = np.linspace(min(sample_means), max(sample_means), 1000)
plt.plot(x, norm.pdf(x, population_mean, population_std / np.sqrt(sample_size)), 
         color='red', linestyle='--', label='Normal Curve (CLT Prediction)')

plt.title('Distribution of Sample Means (Central Limit Theorem)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# 14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
#distribution
import numpy as np
from scipy.stats import norm

# Sample dataset
data = [10, 12, 15, 18, 20, 22, 24, 26, 30]

# Data point to evaluate
x = 22

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data, ddof=0)  # Population standard deviation

# Calculate Z-score
z_score = (x - mean) / std_dev

# Compare to standard normal distribution
# Calculate cumulative probability for Z-score
probability = norm.cdf(z_score)

print(f"Data point: {x}")
print(f"Mean of data: {mean:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"Probability of a value <= {x} in standard normal distribution: {probability:.4f}")
```

```python
# 15.Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

# Sample data
sample = [102, 98, 101, 100, 105, 99, 97, 103, 106, 100]

# Known population parameters
population_mean = 100
population_std = 15
alpha = 0.05

# Sample statistics
sample_mean = np.mean(sample)
n = len(sample)

# Z-test statistic
z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z)))

# Output results
print(f"Sample Mean = {sample_mean}")
print(f"Z-statistic = {z:.4f}")
print(f"P-value = {p_value:.4f}")

# Decision
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# 16.Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy import stats

# Sample data (exam scores)
data = [78, 85, 92, 88, 76, 95, 89, 84, 91, 87]

# Step 1: Calculate the mean and standard error
mean = np.mean(data)
sem = stats.sem(data)  # Standard Error of the Mean

# Step 2: Compute the 95% confidence interval
confidence = 0.95
n = len(data)
margin = sem * stats.t.ppf((1 + confidence) / 2.0, n - 1)
ci_lower = mean - margin
ci_upper = mean + margin

print(f"Mean = {mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Generate 1000 data points from N(50, 10^2)
data = np.random.normal(loc=50, scale=10, size=1000)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample std deviation
n = len(data)
standard_error = sample_std / np.sqrt(n)


import scipy.stats as stats

# 95% confidence level
confidence_level = 0.95
z_critical = stats.norm.ppf((1 + confidence_level) / 2)

# Margin of error
margin_of_error = z_critical * standard_error

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# 21.Write a Python program to compare the standard deviations of two datasets and visualize the difference
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Datasets (you can change these values)
data1 = [12, 15, 14, 10, 18, 20, 22, 25, 21]
data2 = [10, 13, 11, 10, 9, 8, 12, 13, 11]

# Calculate standard deviations
std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

# Compare standard deviations
if std1 > std2:
    print("Dataset 1 has higher variability.")
elif std2 > std1:
    print("Dataset 2 has higher variability.")
else:
    print("Both datasets have equal variability.")

# Visualization
plt.figure(figsize=(10, 6))
sns.kdeplot(data1, fill=True, color='blue', label='Dataset 1')
sns.kdeplot(data2, fill=True, color='green', label='Dataset 2')
plt.axvline(np.mean(data1), color='blue', linestyle='--', label='Mean of Dataset 1')
plt.axvline(np.mean(data2), color='green', linestyle='--', label='Mean of Dataset 2')
plt.title('Comparison of Distributions and Standard Deviations')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
#distribution.
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution(size=1000, mean=0, std_dev=1):
    """
    Generates a dataset from a normal distribution and calculates its skewness and kurtosis.

    Parameters:
    - size (int): Number of data points to generate.
    - mean (float): Mean of the normal distribution.
    - std_dev (float): Standard deviation of the normal distribution.

    Returns:
    - dict: A dictionary containing skewness and kurtosis.
    """
    # Generate data from a normal distribution
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # Fisherâ€™s definition (normal ==> 0)

    return {
        "Skewness": data_skewness,
        "Kurtosis": data_kurtosis
    }

# Example usage
result = analyze_distribution()
print("Skewness:", result["Skewness"])
print("Kurtosis:", result["Kurtosis"])
```
